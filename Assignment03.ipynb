```python
#out of the top 100 billionaires, see how many are male vs nonmale
count = 0
male = 0
nonmale = 0

#4. for loop
for index, row in myDf.iterrows():
    #3. conditional expression
    if count >= 100:
        break
    if row['demographics.gender'] == "male":
        #2. mathematical operation
        male += 1
    else:
        nonmale += 1
    count += 1
    
#print(male, nonmale) 

#5. function definition: compare men vs non-men billionaires
def makeOutput(numMale, numNonMale):
    return ("Out of the top 100 billionaires, there are " + str(numMale) + 
            " male and " + str(numNonMale) + " non-male.")

#5. function execution
print(makeOutput(male, nonmale))
```

```python
mydict = {"gender": ["male", "non-male"],
            "count": [male, nonmale]
    }

f = pd.DataFrame(mydict)
#print(f)
e = f.plot(kind="pie", y="count", labels=f["count"], title="How many of the top 100 billionaires are men?")
e.legend(f["gender"])
```

```python
e = f.plot(kind="bar", x="gender", y="count", title="How many of the top 100 billionaires are men?")
```

As we can see from the charts above, 82 of the top 100 billionaires are men. These two plots above make it much easier to visualize the gender disparity.
